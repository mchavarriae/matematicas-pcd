Demuestre de manera matemática si los siguientes sistemas $L\left\{ \cdot\right\}$  (con entrada $u(t)$ y salida $g(t)$ ) son lineales o no lineales (**escriba las fórmulas en celdas de texto**). Una vez hecho su mayor esfuerzo, si no sabe cómo seguir matemáticamente, puede sustituir con valores y mostrar por contra-ejemplo.

a)  
$g\left(t\right)=\log _{10} u(t)$







Para que cumpla linealidad debe probarse que se cumplan las propiedades de homogeneidad y superposición. Para esto debe cumplirse:
$$L\left\{ \alpha f_{1}\left(x\right)+\beta f_{2}\left(x\right)\right\} =\alpha L\left\{ f_{1}(x)\right\} +\beta L\left\{ f_{2}(x)\right\}$$

Partimos de las siguientes propiedades de logaritmos:

$\log _{a}b^n=\ n * log _{a} b$

$\log _{a} b * c =\log _{a} b + \log _{a} c $

En este ejercicio:

$$log _{10}\left\{ \alpha u_{1}\left(t\right)+\beta u_{2}\left(t\right)\right\} =?\alpha\log _{10}\left\{u_{1}\left(t\right)\right\}+\beta\log _{10}\left\{u_{2}\left(t\right)\right\}$$

$$log _{10}\left\{ \alpha u_{1}\left(t\right)+\beta u_{2}\left(t\right)\right\} =?log _{10}\left\{u_{1}\left(t\right)^\alpha\right\}+log _{10}\left\{u_{2}\left(t\right)^\beta\right\}$$

$$log _{10}\left\{ \alpha u_{1}\left(t\right)+\beta u_{2}\left(t\right)\right\} =?log _{10}\left\{u_{1}\left(t\right)^\alpha\ * u_{2}\left(t\right)^\beta\right\}$$

Para que esto sea igual debe cumplirse que:
$$\alpha u_{1}\left(t\right)+\beta u_{2}\left(t\right)\=?u_{1}\left(t\right)^\alpha\ * u_{2}\left(t\right)^\beta\$$

Podemos usar a manera de contra ejemplo:
$$u_{1}\left(t\right)=u_{2}\left(t\right)\ =1$$
$$\alpha=\beta=2$$
Si sustituimos sería:
$$\left(2\right)\left(1\right)+\left(2\right)\left(1\right)=?\left(1\right)^2 * \left(1\right)^2$$
$$4\neq1$$

Por tanto no se cumple linealidad

b)  
$g\left(t\right)= 5 * u(t) + 13$



Para demostrar matemáticamente si el sistema es lineal, aplicamos la propiedad de linealidad condensada (Superposición + Homogeneidad). El sistema debe cumplir la siguiente igualdad para cualesquiera entradas $u_1(t), u_2(t)$ y escalares $\alpha, \beta$:

$$L\{\alpha \cdot u_1(t) + \beta \cdot u_2(t)\} = \alpha \cdot L\{u_1(t)\} + \beta \cdot L\{u_2(t)\}$$

*Paso 1: Calcular el lado izquierdo (Aplicar el sistema a la entrada combinada)*
Sustituimos la entrada $u(t)$ en la ecuación del sistema por la combinación lineal $(\alpha \cdot u_1(t) + \beta \cdot u_2(t))$:

$$L\{\alpha \cdot u_1(t) + \beta \cdot u_2(t)\} = 5 \cdot (\alpha \cdot u_1(t) + \beta \cdot u_2(t)) + 13$$
$$= 5 \alpha u_1(t) + 5 \beta u_2(t) + 13$$

*Paso 2: Calcular el lado derecho (Combinación lineal de las salidas individuales)*
Calculamos la salida para cada entrada por separado y ponderamos por $\alpha$ y $\beta$:

$$L\{u_1(t)\} = 5 u_1(t) + 13$$
$$L\{u_2(t)\} = 5 u_2(t) + 13$$

Ahora multiplicamos por los escalares y sumamos:
$$\alpha \cdot L\{u_1(t)\} + \beta \cdot L\{u_2(t)\} = \alpha \cdot (5 u_1(t) + 13) + \beta \cdot (5 u_2(t) + 13)$$
$$= 5 \alpha u_1(t) + 13 \alpha + 5 \beta u_2(t) + 13 \beta$$
$$= 5 \alpha u_1(t) + 5 \beta u_2(t) + 13(\alpha + \beta)$$

*Paso 3: Comparación y Conclusión*
Comparamos los resultados obtenidos en el Paso 1 y el Paso 2:

$$5 \alpha u_1(t) + 5 \beta u_2(t) + \mathbf{13} \quad \neq \quad 5 \alpha u_1(t) + 5 \beta u_2(t) + \mathbf{13(\alpha + \beta)}$$

Observamos que los términos constantes difieren. La igualdad no se cumple para cualquier valor de $\alpha$ y $\beta$ (por ejemplo, si $\alpha=2$ y $\beta=2$, tendríamos $13 \neq 52$).

a) Proponga dos vectores:  $x$  e  $y$ que sean colineales (con dos elementos cada uno).  Programe en python para mostrar que son colineales y luego grafíquelos en un mismo gráfico en python. En el cuaderno visto en clase viene un ejemplo de uso de import matplotlib.pyplot as plt



Dos vectores colineales:  
Definimos un vector $x$ (cualquiera) de dos elementos y un vector $y = \alpha x$  donde $\alpha=3$  ($\alpha$ podría ser cualquier escalar)
$x=\begin{bmatrix}2 \\ 5 \end{bmatrix}, y=\begin{bmatrix}6 \\ 15 \end{bmatrix}$  
Ahora probamos colinealidad:

```python
#Definiendo los vectores
x = np.array([2,5])
y = 3 * x
print( x )
print( y )

# probando colinealidad

#Calculando el producto cruzado, deberia ser 0
p_cruzado = x[0]*y[1] - x[1]*y[0]
print(p_cruzado)

if p_cruzado == 0:
  print("Son colineales")
else:
  print("No son colineales")
```

Graficamos los vectores:

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA
import math
%matplotlib inline

# graficar
eX = np.array([0.,0.])
eY = np.array([0.,0.])
x = np.array([2,5])
y = 3 * x


fig, ax = plt.subplots()
ax.quiver(eX, eY, x, y,color=["r","b"],angles='xy', scale_units='xy', scale=1)
ax.axis([-5.,20.,-5.,20.])
ax.set_aspect('equal')
plt.show()

a = np.array( [x[0], y[0] ] )
b = np.array( [x[1], y[1] ] )
cosFI = np.dot(a,b) / ( np.dot( LA.norm(a), LA.norm(b) ))
print("coseno del ángulo_Fi: ",  cosFI )
radianes = math.acos(cosFI)
print("En Radianes el ángulo_Fi: ", radianes)
grados = math.degrees( radianes )
print("En Grados el ángulo_Fi: ", grados )
```

En el archivo "reales.csv" se encuentran los valores reales de un conjunto de datos, y en el archivo "predicciones.csv" lo que predijo un algoritmo.

Cargue ambos archivos y muestre la norma 2 y la norma 5, de la diferencia entre el real y el predicho.

Finalmente, grafique el predicho y el real en un mismo gráfico para comparar.

```python
# carga archivos

archivo = "reales.csv"
dataFrameReales = pd.read_csv(archivo, header = 0)

archivo = "predicciones.csv"
dataFramePredichos = pd.read_csv(archivo, header = 0)


# Convertir de pandas a numpy
reales = dataFrameReales.to_numpy().flatten()
predichos = dataFramePredichos.to_numpy().flatten()


# Calcular las normas
l2 = np.linalg.norm(reales - predichos, ord=2)
l5 = np.linalg.norm(reales - predichos, ord=5)

print("Norma L2: ", l2)
print("Norma L5: ", l5)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(reales, label='Valores Reales', color='blue')
plt.plot(predichos, label='Predicciones', color='orange', alpha=0.7) # alpha da transparencia
plt.legend()
plt.title("Comparación: Reales vs Predichos")
plt.show()
```
